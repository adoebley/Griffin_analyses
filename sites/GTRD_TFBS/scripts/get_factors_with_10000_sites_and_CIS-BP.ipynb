```python
import os
import pandas as pd
```

```python
#path to directory with bed files
in_dir='../filtered_sites/sites/'

#minimum number of sites
min_sites = 10000

#path to output yaml
out_file='all_TFBS_'+str(min_sites)+'_sites_CIS-BP.yaml'
out_dir = '../'+str(min_sites)+'_filtered_sites_CIS_BP/'

#path to cis-bp info
CIS_BP_list = '../CIS_BP/Homo_sapiens_2021_04_29_Cis-BP_v_2.00/TF_Information.txt'
```

```python
valid_TFs = pd.read_csv(CIS_BP_list, sep='\t')#['TF_Name']
valid_TFs = valid_TFs.sort_values(by = 'TF_Name')
#exclude TFs where the binding site is not known
valid_TFs = valid_TFs[valid_TFs['TF_Status']!='N']
#get just TF names
valid_TFs = pd.Series(valid_TFs['TF_Name'].unique())
valid_TFs = valid_TFs+'.hg38'
```

```python
chroms = ['chr'+str(m) for m in range(1,23)]
```

```python
out_dir=os.path.abspath(out_dir)+'/'
output=[]
for item in sorted(os.listdir(in_dir)):
    if item.startswith('._'):
        continue
    elif item.endswith('.high_mapability.txt'):
        name = item.split('.high_mapability.txt')[0]
        data = pd.read_csv(in_dir+item, sep='\t')
        #only keep autosomes
        data = data[data['Chrom'].isin(chroms)]
        #get the top sites
        data = data.sort_values(by='peak.count', ascending=False).iloc[0:min_sites]
        if len(data)==min_sites and name in valid_TFs.values: #if there were at least min_sites on the list and it is a real TF
            output.append([name,out_dir+name+'.'+str(min_sites)+'.txt'])
            data.to_csv(out_dir+name+'.'+str(min_sites)+'.txt', sep='\t', index=False)
        else: 
            print(name,len(data))
```

```python
print(len(output))
```

```python
with open (out_file, 'w+' )as f:
    f.write('site_files:\n  ')
    f.write('\n  '.join([': '.join(item) for item in output]))
```
