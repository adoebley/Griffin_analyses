```python
import pandas as pd
```

```python
in_dict = {}
for correction in ['single','multi']:
    for size in ['35-100bp','100-200bp']:
        current = pd.read_csv('files/'+correction+'_fragment_length_'+size+'.tsv', sep='\t').set_index('site_name')
        in_dict[correction+'.'+size] = current
```

```python
in_dict[correction+'.'+size]
```

```python
merged = pd.DataFrame(index = in_dict[correction+'.'+size].index)

for correction in ['single','multi']:
    for size in ['35-100bp','100-200bp']:
        current = in_dict[correction+'.'+size].copy()
        current.columns = current.columns+'.'+correction+'_length.'+size
        merged[current.columns] = current[current.columns]
```

```python
merged.to_csv('files/S2_single_vs_multi_length_GC.tsv',sep='\t')
```
