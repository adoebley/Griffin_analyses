```python
import pysam
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
midpoint = pd.read_csv('../../midpoints_vs_coverage/results/HD45.ctDNA.WGS.FC19269447/HD45.ctDNA.WGS.FC19269447.GC_corrected.coverage.midpoints.tsv', sep='\t')
```

```python
fragments = pd.read_csv('../../midpoints_vs_coverage/results/HD45.ctDNA.WGS.FC19269447/HD45.ctDNA.WGS.FC19269447.GC_corrected.coverage.fragments.tsv', sep='\t')
```

```python
plot_columns = np.arange(-990,990,15)
str_plot_columns = [str(m) for m in plot_columns]
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(plot_columns, midpoint[str_plot_columns].T.values, label = 'Midpoint coverage', color = 'green')
ax.plot(plot_columns, fragments[str_plot_columns].T.values, label = 'Fragment coverage', color = 'black')

ax.legend(bbox_to_anchor = [1,1])
ax.set_title('LYL1 sites')
ax.set_xlabel('Distance from site')
ax.set_ylabel('Nrmalized coverage')
ax.set_aspect(1.0/ax.get_data_ratio(), adjustable='box')
fig.tight_layout()
fig.savefig('plots/midpoint_vs_coverage.pdf')
```
